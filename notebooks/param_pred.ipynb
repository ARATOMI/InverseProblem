# Partly independent model with probability distribution prediction

One shared MLP and 22 independent models (11 for mean and 11 for standard deviation)

```python
import json
from inverse_problem.nn_inversion.main import HyperParams, Model
from inverse_problem.nn_inversion.posthoc import compute_metrics, open_param_file, plot_params, plot_2d, plot_1d, plot_param_hist
from pprint import pprint
from astropy.io import fits
import numpy as np
```

### Load data

```python
def create_mask(data):
    boundaries = np.array([(np.min(data[:, j]) + 10**(-6), np.max(data[:, j]) - 10**(-6)) for j in range(11)])
    pop_inds = []
    mask = np.zeros(data.shape)
    for i in range(data.shape[0]):
        if sum([int(data[i, j] < boundaries[j, 0]) + int(boundaries[j, 1] < data[i, j]) for j in range(11)]):
            pop_inds.append(i)
            mask[i] = np.ones(11)
    return(np.array(pop_inds)), mask
```

```python
path_to_json = '../res_experiments/hps_partly_independent_dist_mlp_mask.json'
with open(path_to_json) as f:
    json_data = json.load(f)
```

```python
refer_path = '../data/reference.fits'
refer, names = open_param_file(refer_path)
_, mask = create_mask(refer.reshape(-1, 11))
masked_refer = np.ma.array(refer.reshape(-1, 11), mask = mask)
masked_refer = masked_refer.reshape(512, 485, 11)
```

```python
refer_unnorm, _ = open_param_file(refer_path, normalize=False)
_, mask = create_mask(refer_unnorm.reshape(-1, 11))
masked_refer_unnorm = np.ma.array(refer_unnorm.reshape(-1, 11), mask = mask)
masked_refer_unnorm = masked_refer_unnorm.reshape(512, 485, 11)
```

```python
hps = HyperParams.from_file(path_to_json=path_to_json)
hps
```

### Train model on a small part of dataset


```python
params = fits.open('../data/small_parameters_base.fits')[0].data
pop_inds, _ = create_mask(params)
params = np.delete(params, pop_inds, axis=0)
```

```python
model = Model(hps)
```

```python
history = model.train(data_arr=params)
```

```python
predicted_mu, predicted_sigma = model.predict_refer(masked_refer_unnorm)
```

```python
df_ind = compute_metrics(masked_refer, predicted_mu, names, True)
df_ind
```

### Train model on the entire dataset

```python
params_full = fits.open('../data/parameters_base.fits')[0].data
pop_inds, _ = create_mask(params_full)
params_full = np.delete(params_full, pop_inds, axis=0)
```

```python
full_model = Model(hps)
```

```python
print("Number of training parameters:", sum(p.numel() for p in full_model.net.parameters() if p.requires_grad))
```

```python
history = full_model.train(
    data_arr=params_full,
    logdir = '../logs/',
    path_to_save='../models/pi_dist_mlp_mask_3.pt')
```

### Load the trained model

```python
full_model = Model(hps)
full_model.load_model('../models/pi_dist_mlp_mask_3.pt')
```

```python
full_predicted_mu, full_predicted_sigma = full_model.predict_refer(masked_refer_unnorm)
```

```python
full_df_ind = compute_metrics(masked_refer, full_predicted_mu, names, True, save_path = '../pi_dist_pred_mask_3.csv')
full_df_ind
```
