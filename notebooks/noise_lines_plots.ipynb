## Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from inverse_problem.milne_edington.me import HinodeME, me_model, _compute_spectrum, _prepare_base_model_parameters, _prepare_zero_model_parameters
```

```python
param_vec = [1000., 15., 20., 30., 1., 50., 0.5, 0.5, 0, 0.7, -9]
```

```python
hinode = HinodeME(param_vec)
noised = hinode.compute_spectrum()
```

```python
def plot_lines(lines):
    line_type = ['I','Q','U','V']
    fig, ax = plt.subplots(2,2, figsize = (10,5))
    for i in range(4):
        ax[i//2][i%2].plot(lines[:, i]); ax[i//2][i%2].set_title(f'Spectral line {line_type[i]}')
    fig.set_tight_layout(tight = True)

a = plot_lines(noised)
```

```python
no_noise = hinode.compute_spectrum(with_noise=False)
plot_lines(no_noise)
```
